## Step 1: Upload Project Files

Run this cell, then upload your project ZIP file when prompted.

```python
from google.colab import files
import zipfile
import os

print("📁 Please upload your ORION-DDH-RESISITIVITY.zip file...")
uploaded = files.upload()

# Extract the zip
for filename in uploaded.keys():
    print(f"\n📦 Extracting {filename}...")
    with zipfile.ZipFile(filename, 'r') as zip_ref:
        zip_ref.extractall('/content/')
    print("✅ Extraction complete!")

# Find the project directory
for item in os.listdir('/content/'):
    if 'ORION' in item.upper() and os.path.isdir(f'/content/{item}'):
        PROJECT_DIR = f'/content/{item}'
        break
else:
    PROJECT_DIR = '/content/ORION-DDH-RESISITIVITY'

print(f"\n📂 Project directory: {PROJECT_DIR}")
%cd {PROJECT_DIR}
!ls -la
```

## Step 3: Build the APK 🔨

This is the main build step. It will take 30-45 minutes on first run.

```python
# Build APK
import os
os.environ['BUILDOZER_WARN_ON_ROOT'] = '0'

print("🔨 Starting APK build... This will take 30-45 minutes.")
print("☕ Grab a coffee!\n")

!yes | buildozer android debug 2>&1 | tail -100

print("\n✅ Build process completed!")
```

## Step 4: Download Your APK 📱

```python
import glob
from google.colab import files

# Find the APK
apk_files = glob.glob('bin/*.apk')

if apk_files:
    apk_path = apk_files[0]
    print(f"🎉 APK built successfully!")
    print(f"📁 File: {apk_path}")
    print(f"📏 Size: {os.path.getsize(apk_path) / 1024 / 1024:.1f} MB")
    print("\n📥 Downloading...")
    files.download(apk_path)
else:
    print("❌ APK not found. Check build logs above for errors.")
    print("\nLooking for build output:")
    !ls -la bin/ 2>/dev/null || echo "bin/ directory not found"
```

## 📱 Installing on Android

1. Transfer the APK to your Android device
2. Go to **Settings → Security → Install unknown apps**
3. Allow your file manager/browser to install apps
4. Tap the APK file to install
5. Open **Orion-DDH_v1** - look for the Doraemon icon! 🤖
